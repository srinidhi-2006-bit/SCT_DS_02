```python
#Exploratory Data Analysis on Titanic Survivors
```

```python
# Step-01:Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Step-02:Load the dataset
df=pd.read_csv('C:\\Users\\egala\\OneDrive\\Desktop\\Data_Science_Internship_at_skillcraft\\train.csv')
```

```python
#Step-03: View structure
print(df.shape)
print(df.columns)
print(df.head())
```

```python
#Step-04:Check missing values(if any)
print(df.isnull().sum())
```

```python
#Step-05:Clean the data
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
#Step-06:Drop the unused columns
df.drop(['Cabin', 'Ticket', 'Name'], axis=1, inplace=True)
```

```python
#Step-07:Convert categorical to numeric
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
#Step-08:Visualizations
```

```python
# Survival by Gender
sns.countplot(data=df, x='Survived', hue='Sex', palette='Set2')
plt.title('Survival by Gender')
plt.show()
```

```python
# Survival by Class
sns.countplot(data=df, x='Survived', hue='Pclass', palette='cool')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data=df[df['Survived'] == 1], x='Age', color='green', label='Survived', kde=True)
sns.histplot(data=df[df['Survived'] == 0], x='Age', color='red', label='Did Not Survive', kde=True)
plt.legend()
plt.title('Age Distribution of Survivors vs Non-survivors')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='icefire')
plt.title('Correlation Between Titanic Features')
plt.show()
```
